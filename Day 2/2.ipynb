```python
f = open("../Data/data2.txt", "r")
data = f.read()
data_clean = data.split("\n")
```

```python
data_clean
```

```python
depth = 0
position = 0
aim = 0

for command in data_clean:
    command_txt = command.split(" ")[0]
    command_num = int(command.split(" ")[1])
    if command_txt == "forward":
        position += command_num
        depth += aim * command_num
    elif command_txt == "down":
        aim += command_num
    elif command_txt == "up":
        aim -= command_num
        
        
print(depth * position)
    
```
